```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Enable inline plots
%matplotlib inline
```

```python
# 4x4 exponential random array
exp_array = np.random.exponential(scale=1.0, size=(4, 4))
print("4x4 Exponential Array:")
print(exp_array)
```

```python
# Generate distributions
data_exp = np.random.exponential(size=100000)
data_uniform = np.random.rand(100000)
data_normal = np.random.randn(100000)

# Plot histograms
plt.figure(figsize=(10, 5))
plt.hist(data_exp, density=True, bins=150, histtype="step", label="Exponential", color="red")
plt.hist(data_uniform, density=True, bins=150, histtype="step", label="Uniform", color="green")
plt.hist(data_normal, density=True, bins=150, histtype="step", label="Normal", color="blue")

plt.title("Comparison of Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.axis([-2.5, 10, 0, 1])
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title('Z = X + Y')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

# Select relevant features
features = ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]
selected_df = df[features]

# Pearson Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(selected_df.corr(method='pearson'), annot=True, cmap='Blues')
plt.title("Pearson Correlation Heatmap")
plt.show()

# Spearman Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(selected_df.corr(method='spearman'), annot=True, cmap='Greens')
plt.title("Spearman Correlation Heatmap")
plt.show()
```
